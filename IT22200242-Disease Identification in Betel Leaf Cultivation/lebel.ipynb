```python
folder = 'Data segrigation/Bacteria Blight/'
count = 1

for filename in os.listdir(folder):
    source = folder + filename
    destination = folder + "Bacteria Blight_" +str(count)+".jpg"
    os.rename(source, destination)
    count+=1
print("All files are renamed in the Bacteria Blight dir.")
```
